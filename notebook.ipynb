## Visualization & Explanatory Analysis

### Pertanyaan 1: 

Berapa banyak sepeda disewakan dalam satu hari?

```python
penyewaan = hari.groupby('weekday')['cnt'].mean()
```

```python
penyewaan
```

```python
penyewaan.plot(kind='line',figsize=(14,5))

plt.xlabel('Hari')
plt.ylabel('Penyewaan')
plt.title('Penyewaan sepeda dalam perhari')

plt.show()
```

### Pertanyaan 2: 

Berapa banyak penyewaan sepeda dalam perbulan?

```python
bulan = hari.groupby('mnth')['cnt'].sum()
```

```python
bulan
```

```python
bulan.plot(kind='bar',figsize=(14,5))

plt.xlabel('Bulan')
plt.ylabel('Penyewaan')
plt.title('Penyewaan sepeda dalam perbulan')

plt.show()
```
